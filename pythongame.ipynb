Python game

```python
import random

secret_num=random.randint(1,100)

for i in range(0,secret_num+1):
  user_input=int(input("guess the number:"))
  if secret_num < user_input:
    print("Too High! , try again")
  elif secret_num > user_input:
    print("Too Low! , try again")
  else:
    print("Correct.....")



```
