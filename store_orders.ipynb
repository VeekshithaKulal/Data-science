Basic python functions


```python
import pandas as pd
```

```python
df=pd.read_csv("/content/store_orders - pandas 3.csv")
```

```python
print(df.to_string())
```

```python
# Load the dataset and convert Order_Date to datetime.
df=df.rename(columns={"Order_Date":"datetime"})
```

```python
print(df.to_string())
```

```python
# Create a new column "Total_Amount" = Quantity Ã— Price
df["Total_Amount"]=df["Quantity"]*df["Price"]
```

```python
print(df.to_string())
```

```python
# Identify top 3 highest-value orders.
df.sort_values(by='Total_Amount',ascending=False).head(3)
```

```python
# Group by Category and show total revenue.
df.groupby("Category")["Total_Amount"].sum()
```

```python
# Group by City and find number of orders from each city.
df.groupby("City")["Quantity"].sum()
```

```python
# Find which customer spent the most money overall.
df.groupby("Customer_Name")["Price"].sum().sort_values(ascending=False).head(1)
```

```python
# Extract all orders during May 2024 (using query / datetime filtering)
df.query("datetime>='2024-05-01' and datetime <='2024-05-31'")
```

```python
# Detect and remove duplicate orders (if any).
# remove (df.drop_duplicates())
df.duplicated().any()
```

```python
# Find revenue trend per day (daily sales).
df.groupby("datetime")["Total_Amount"].sum()
```

```python
# Add a column:

# If Total_Amount > 20,000 â†’ "High Value"

# else â†’ "Normal"

df["prediction"]=df["Total_Amount"].apply(lambda x: "High value" if x >20000 else "normal")
print(df.to_string())
```
