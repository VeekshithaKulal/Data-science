Basic numpy



```python
import numpy as np
```

```python
# Create a NumPy array from a list or tuple.
arr = np.array([1, 2, 3])
print(arr)
```

```python
# Purpose: Creates range of values (start, stop, step).
np.arange(1, 10, 2)
```

```python
# Purpose: Creates evenly spaced values between start & stop.

np.linspace(0, 1, 5)
```

```python
# Purpose: Creates array filled with zeros.
np.zeros((3, 3))
```

```python
# Purpose: Creates array filled with ones.
np.ones(5)
```

```python
# Purpose: Identity matrix.
np.eye(4)
```

```python
# Purpose: Change the shape of an array without changing data.
s=np.array([1,9,5,8,3,2])
s.reshape(2, 3)
```

```python
# Purpose: Adds all elements.
np.sum(arr)
```

```python
# Purpose: Calculates average value.
np.mean(arr)
```

```python
#  Purpose: Computes standard deviation.
np.std(arr)
```

```python
# Purpose: Finds smallest or largest value.
# np.max(arr)
np.min(arr)
```

```python
# Generates random numbers between 0 and 1.
np.random.rand(3)
```

```python
# Random integers in a specified range.
np.random.randint(1, 100)
```

```python
# Combines arrays.
a=np.array([1,2,3,4])
b=np.array([6,7.8,9])
np.concatenate([a, b])
```

```python
# Purpose: Returns unique elements.
np.unique(arr)
```

```python
# Sorts an array.
np.sort(arr)
```

```python
#  Access specific rows, columns and slices.
arr[0]
# arr[:, 1]
```

```python
# Returns positions where condition is true.
np.where(arr > 10)
```

```python
# Purpose: Matrix multiplication / dot product.
a=np.array([1,2,3,4])
b=np.array([6,7,8,9])
np.dot(a, b)

```

```python
# Purpose: Swaps rows and columns.
arr.T
```
